```python
df.drop(['Unnamed: 32', 'id'],  axis = 1, inplace=True) 
df.head()
```

```python
df['diagnosis'] = df['diagnosis'].map({'M' : 1, 'B' : 0})
df.head()
```

```python
x = df.drop('diagnosis', axis=1)
x.columns
```

```python
y = df['diagnosis']
y
```

```python
from sklearn.utils import shuffle
x, y = shuffle(x, y)
```

```python
#scaling [-1, 1]  MinMaxScaler
from sklearn import preprocessing
x_MinMaxScaler = preprocessing.minmax_scale(x,feature_range=(-1,1))
x_MinMaxScaler
```

```python
x_robust = preprocessing.robust_scale(x)
x_robust
```

```python
# #split test/train
# from sklearn.model_selection import train_test_split
# x_train, x_test, y_train, y_test = train_test_split(x_MinMaxScaler, y, test_size=0.2)
```

```python
from sklearn.svm import SVC
model = SVC(C=0.3,kernel="linear")
```

```python
import time
from sklearn.model_selection import RandomizedSearchCV
parameters={'C':[0.1,0.01,0.2,0.3,0.5,1],
            'kernel':['linear','poly','rbf','sigmoid'],
            }
tic=time.time()
Randomized_SearchCV=RandomizedSearchCV(estimator=model,param_distributions=parameters,n_iter=20)
Randomized_SearchCV.fit(x, y)
print(Randomized_SearchCV.best_score_ *100)
toc=time.time()
print(toc-tic)
```

```python
Randomized_SearchCV.best_params_
```

```python
from sklearn.model_selection import cross_val_score
import time
k_fold =10
tic=time.time()
k_score=cross_val_score(model,x_MinMaxScaler,y,cv=k_fold)
toc=time.time()
print(k_score)
print("K_fold Average of Accuracy = ",k_score.mean()*100,"%")
print('time:  ',toc-tic)
```

```python
from sklearn.model_selection import cross_val_predict
y_pred = cross_val_predict(model, x_MinMaxScaler,y, cv=10)
```

```python
import seaborn as sn
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(y, y_pred)
sn.heatmap(confusion_matrix, annot=True, fmt=".4g",xticklabels=['B','M'],yticklabels=['B','M']) # font size
```
